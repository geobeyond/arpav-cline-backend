```python
for cov_conf in database.collect_all_coverage_configurations(session):
    print(cov_conf.name)
    if (maximum := cov_conf.color_scale_max) - (minimum := cov_conf.color_scale_min) < 0.00001:
        print("\n\n\nSKIPPING\n\n\n")
    else:
        parsed_palette = palette.parse_palette(cov_conf.palette, settings.palettes_dir)
        print(parsed_palette)
        applied_palette = palette.apply_palette(parsed_palette, cov_conf.color_scale_min, cov_conf.color_scale_max, num_stops=5) 
        print(applied_palette)
    print("-----")
```
